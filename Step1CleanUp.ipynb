```python
import pandas as pd
import requests
from citipy import citipy
from config import weather_api_key
from datetime import datetime
```

```python
TodayDate = '5/28/20'
df0 = pd.read_csv("time_series_covid19_confirmed_global.csv")
df0
```

```python
df1 = df0[['Province/State','Country/Region','Lat','Long',TodayDate]]
df1['Date'] = TodayDate
df1 = df1.rename(columns={TodayDate: 'Cases'})
df1
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
for i, row in df1.iterrows():
    lat = row['Lat']
    lng = row['Long']
    city = citipy.nearest_city(lat, lng).city_name
    df1.at[i, 'City'] = city
    #df3.set_value(i, 'City', city)
    print(str(lat) + ', ' + str(lng) + ': ' + city)
```

```python
df4 = df1.copy()
df4
```

```python
count = 0
for i, row in df4.iterrows():
    city = row['City']
    city_url = url + "&q=" + city
    try:
        city_weather = requests.get(city_url).json()
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        try:
            city_rain = city_weather["rain"]["3h"]
        except:
            city_rain = 0
        try:
            city_snow = city_weather["snow"]["3h"]
        except:
            city_snow = 0
        df4.at[i, 'Temp_Max'] = city_max_temp
        df4.at[i, 'Humidity'] = city_humidity
        df4.at[i, 'Rain'] = city_rain
        df4.at[i, 'Snow'] = city_snow
    except:
        print("City not found.")
        count += 1
    #df3.set_value(i, 'City', city)
    #print(str(lat) + ', ' + str(lng) + ': ' + city)
print(str(count) + ' cities not found')
df4
```

```python
#df4.to_csv(r'ProcessedData.csv')
df2 = pd.read_csv('ProcessedData.csv')
del df2['Unnamed: 0']
df2
```

```python
frames = [df2, df4]
df3 = pd.concat(frames)
df3
```

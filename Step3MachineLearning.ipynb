```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder


from sqlalchemy import create_engine
import psycopg2 #pip install psycopg2
from config import db_password

DBEnabled = True
```

```python
if not DBEnabled:
    print('Database disabled')
    df = pd.read_csv("ProcessedData.csv")
else:
    print('Database enabled')
    str_db = f"postgres://postgres:{db_password}@127.0.0.1:5432/Team7"
    
    engine = create_engine(str_db)
    
    df = pd.read_sql_query('select * from "Team7"',con=engine)
df.head()
```

```python
df = df.dropna()
df
```

```python
# Bucketing eventually in next segments
df = df.drop(["Unnamed: 0", "Province/State", "Lat", "Long", "Date"], 1)
y = df["Cases"]
x = df.drop(["Cases"], 1).values
```

```python
list_categories = df.dtypes[df.dtypes == "object"].index.tolist()
#Encoder
enc = OneHotEncoder(sparse=False)

encode_df = pd.DataFrame(enc.fit_transform(df[list_categories]))
encode_df.columns = enc.get_feature_names(list_categories)
encode_df.head()
```

```python
df = df.merge(encode_df, left_index=True, right_index=True).drop(list_categories,1)
df.head()
```

```python
X = df.drop(["Cases"], 1).values
y = df["Cases"].values
# Train
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
input_count = len(X_train[0])
hidden_nodes_1 = 20
hidden_nodes_2 = 8
print(len(X_train[0]))
```

```python
# Keras Sequential Model
nn_model = tf.keras.models.Sequential()
nn_model.add(tf.keras.layers.Dense(units=hidden_nodes_1, activation="relu", input_dim=input_count))
nn_model.add(tf.keras.layers.Dense(units=hidden_nodes_2, activation="relu"))
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
nn_model.summary()
```

```python
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```
